Buscar a página principal do UOL (www.uol.com.br) e exibir seu conteúdo no formato texto puro.

_Hint: Beautiful Soup

> Realizar a instalação do módulo bs4 pelo Terminal do Anaconda, digitando __conda install beautifulsoup4 requests__

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://www.uol.com.br"

response = requests.get(url)

soup = BeautifulSoup(response.content, "lxml")
```

### Original

```python
response.content
```

### Pretty

```python
soup.prettify()
```

### Texto puro

```python
soup.text
```
