Utilize o conjunto de dados da challenge anterior, relacione dois atributos em um scatter plot que indique a relação entre os mesmos.

```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('seaborn')
```

```python
df_pimas = pd.read_csv("data/pima-data-norm-city.csv", index_col=0)
df_pimas.head(5)
```

```python
sns.lmplot("age", "num_preg", df_pimas)
```

```python
sns.lmplot("age", "diab_pred", df_pimas)
```

```python
sns.lmplot("age", "glucose_conc", df_pimas)
```

```python
sns.lmplot("diab_pred", "glucose_conc", df_pimas, order=2)
```

```python
sns.lmplot("insulin", "glucose_conc", df_pimas)
```

```python
sns.lmplot("insulin", "diab_pred", df_pimas)
```

```python
sns.lmplot("bmi", "diab_pred", df_pimas)
```
