Utilizando O Dataframe Ajustado Da Challenge Anterior, Exiba As Seguintes Informações:

* Soma De Diabetes Para Idades Entre 20 E 50 Anos
* Soma De Gravidezes Para Idades Até 30 Anos
* Soma De Diabetes Por Estado
* Média da Pressão Diástólica por Cidade

```python
import numpy as np
import pandas as pd
```

```python
df_pimas = pd.read_csv("data/pima-data-norm-city.csv", index_col=0)
df_pimas.head(5)
```

```python
df_filtered = df_pimas[(df_pimas["diabetes"]==True) & (df_pimas["age"] >= 20) & (df_pimas["age"] <=50)]
df_filtered.head(5)
```

```python
group_age = df_filtered["diabetes"].groupby(df_filtered["age"])
group_age.count()
```

```python
df_filtered = df_pimas[(df_pimas["diabetes"]==True) & (df_pimas["age"] <=30)]
df_filtered.head(5)
```

```python
group_age_preg = df_filtered["num_preg"].groupby(df_filtered["age"])
group_age_preg.sum()
```

```python
df_filtered = df_pimas[(df_pimas["diabetes"]==True)]
df_filtered.head(5)
```

```python
group_diab_state = df_filtered["diabetes"].groupby(df_filtered["state"])
group_diab_state.count()
```
