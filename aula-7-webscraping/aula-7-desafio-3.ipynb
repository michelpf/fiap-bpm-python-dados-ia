```python
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=chrome_options)
```

```python
from selenium.webdriver.common.by import By
import time 
```

```python
def get_manchetes(url_root):
  driver.get(url_root)
  time.sleep(3)

  manchetes = driver.find_elements(By.XPATH, '//h3')
  total_manchetes = 0

  for manchete in manchetes:
    if manchete != "":
      classe = manchete.get_attribute(name="class")
      if "title" in classe and manchete.text != "" and len(manchete.text.split(" ")) > 4:
        print(manchete.text)
        total_manchetes += 1

  print("Total de manchetes: " + str(total_manchetes))
```

```python
get_manchetes("https://www.uol.com.br")
```

```python
get_manchetes("https://www.globo.com")
```
